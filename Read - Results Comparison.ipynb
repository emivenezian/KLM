```python
fig = go.Figure()

aircraft_types = {
    '77W': 'B777-300ER',
    '789': 'B787-9',
    '781': 'B787-10',
    '772': 'B777-200ER'
}

combined_results['Aircraft Type'] = combined_results['Aircraft Type'].map(aircraft_types)
# Add scatter plots for each dataset
fig.add_trace(go.Scatter(
    x=combined_results[:-2]['TOW'],
    y=combined_results[:-2]['Fuel Deviation_model'],
    mode='markers',
    name='Model',
    marker=dict(color=colors[0], size = 10)
))

# fig.add_trace(go.Scatter(
#     x=combined_results[:-2]['TOW'],
#     y=combined_results[:-2]['Fuel Deviation_baseline'],
#     mode='markers',
#     name='Baseline',
#     marker=dict(color=colors[1], size = 8)
# ))

# fig.add_trace(go.Scatter(
#     x=combined_results[:-2]['TOW'],
#     y=combined_results[:-2]['Fuel Deviation_optimized_actual'],
#     mode='markers',
#     name='Optimized Actual',
#     marker=dict(color=colors[2], size = 8)
# ))


def add_ellipse_and_annotation(fig, df, aircraft_type):
    subset = df[df['Aircraft Type'] == aircraft_type]
    x_mean = subset['TOW'].mean()
    y_mean = subset['Fuel Deviation_model'].mean()
    x_std = subset['TOW'].std()
    y_std = subset['Fuel Deviation_model'].std()
    
    fig.add_shape(type='circle',
                  xref='x', yref='y',
                  x0=x_mean - 2 * x_std, y0=y_mean - 2 * y_std,
                  x1=x_mean + 2 * x_std, y1=y_mean + 2 * y_std,
                  line=dict(color=colors[0], width=2, dash='dash'))
    
    fig.add_annotation(
        x=x_mean, 
        y=y_mean + 2.2 * y_std,  
        text=aircraft_type,
        showarrow=False,
        font=dict(color=colors[0], size=16)
    )


unique_aircraft_types = combined_results[:-2]['Aircraft Type'].unique()
for idx, aircraft_type in enumerate(unique_aircraft_types):
    add_ellipse_and_annotation(fig, combined_results, aircraft_type)

fig.add_shape(type='line',
              x0=200000,
              y0=0,
              x1=350000,
              y1=0,
              line=dict(color='red', dash='dash'))

# Update layout for better visualization
fig.update_layout(
    title='Fuel Deviation per Aircraft Type',
    xaxis_title='Take off Weight [kg]',
    yaxis_title='Fuel Deviation [kg]',
    template='plotly_white',  
    plot_bgcolor='white', 
    paper_bgcolor='white', 
    font=dict(family='Arial, sans-serif', size=18, color='black'),
    height=1000,
    width=1000,  # Customize font style
    showlegend=True
)

# Show the plot
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
fig.show()
```
